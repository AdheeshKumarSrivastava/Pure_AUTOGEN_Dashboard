```python
class PlanetInfo(BaseModel):
    name:str
    color:str
    distance:str
    size:str
    likeliness_to_earth: str
```

```python
agent_un = AssistantAgent(
    name="text_agent",
    model_client=OllamaChatCompletionClient(model="qwen2.5:7b",response_format=PlanetInfo),system_message="You are a helpful assistant. Answer question accurately",)
```

```python
async def text_message():
    text_msg = TextMessage(content="Provide information about Mars and Earth",source='user')
    result = await agent_un.run(task=text_msg)
    print(result.messages[-1].content)

await text_message()
```
